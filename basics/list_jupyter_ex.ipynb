```python
nested_list = [[1,2,3],
               [4,5,6],
               [7,8,9]]
print(nested_list[0])
# [1, 2, 3]
```

```python
print(nested_list[0][-1])
# 3
```

```python
print(nested_list[0][1])
# 2
```

```python
for inner_list in nested_list:
    print(inner_list)
# [1, 2, 3]
# [4, 5, 6]
# [7, 8, 9]
```

```python
for inner_list in nested_list:
    for ele in inner_list:
        print(ele)
# 1
# 2
# 3
# 4
# 5
# 6
# 7
# 8
# 9
```

```python
main_list = list([1,2,3])
print(main_list)
# [1, 2, 3]
```

```python
# deep copy
xerox_copy = main_list.copy()
xerox_copy.append(4)
print(main_list)
print(xerox_copy)
# [1, 2, 3]
# [1, 2, 3, 4]
```

```python
# shallow copy / ref copy
l1 = list([1,2,3])
l2 = l1
l1[0] = 100
print(l1)
print(l2)
# [100, 2, 3]
# [100, 2, 3]
```

```python
main_copy = [[1,2]]
deep_copy = main_copy
deep_copy[0][0] = 2000
print(main_copy)
# [[2000, 2]]
print(deep_copy)
# [[2000, 2]]
```

```python
data = [1,5,2,4,3]
data.sort()
print(data)
# [1, 2, 3, 4, 5]
data.sort(reverse=True)
print(data)
# [5, 4, 3, 2, 1]
```

```python
print( [x**x for x in range(10)] )
# [1, 1, 4, 27, 256, 3125, 46656, 823543, 16777216, 387420489]
```

```python
print( [x**x for x in range(10) if x%2 !=0] )
# [1, 27, 3125, 823543, 387420489]
```

```python
# unpacking -- term used to assign list elmts to variables without loop
my_list = [10,20,30,40,50]
e1, e2, e3, e4, e5 = my_list
print(e1, e2, e3, e4, e5)
# 10 20 30 40 50
```

```python
# packing
my_list1 = ["one", "two", "three"]
my_list2 = [1, 2 , 3]
my_list3 = list(zip(my_list1, my_list2))
print(my_list3)
# [('one', 1), ('two', 2), ('three', 3)]
print(my_list3[0])
# ('one', 1)
print(type(my_list3[0]))
# <class 'tuple'>
print(type(my_list3))
# <class 'list'>
```

```python
print(len(my_list2))
# 3
print(max(my_list2))
# 3
print(min(my_list2))
# 1
print(sum(my_list2))
# 6
print(len(my_list3))
# 3
```
