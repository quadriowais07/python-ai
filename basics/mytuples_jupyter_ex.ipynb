```python
t1 = (10,20,30,40,50)
print(t1[0])
# 10
print(t1[-5])
# 10
print(t1[0:3])
# (10, 20, 30)
print(t1[:2])
# (10, 20)
print(t1[2:])
# (30, 40, 50)
print(t1[0:4:2])
# (10, 30)
print(t1[::-2])
# (50, 30, 10)
```

```python
t1 = (10, 20, 30, 40, 50)
e1, e2, e3, e4, e5 = t1
print(e1,e2,e3,e4,e5)
# 10 20 30 40 50
e1, *t2, e5 = t1
print(e1,t2,e5)
# 10 [20, 30, 40] 50
t2 = tuple(t2)
print(t2)
# (20, 30, 40)
e2, *t3 = t2
e3, e4 = t3
print(e2,e3,e4)
# 20 30 40
```
