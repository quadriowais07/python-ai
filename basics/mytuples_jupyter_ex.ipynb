```python
t1 = ()
print(t1)
# ()
t2 = (10,20,30,40,50)
print(t2)
# (10,20,30,40,50)
t3 = (10,10.1,"Hello",True,None)
print(t3)
# (10,10.1,"Hello",True,None)
t4 = tuple((10,20,30,40,50))
print(t4)
# (10, 20, 30, 40, 50)
print(type(t1))
# <class 'tuple'>
t5=(10)
print(t5)
# 10
print(type(t5))
# <class 'int'>
t6 = (10,)
print(t6)
# (10,)
print(type(t6))
# <class 'tuple'>
t7=10,20,30
print(t7)
# (10, 20, 30)
print(type(t7))
# <class 'tuple'>
t8=10,
print(t8)
# (10,)
print(type(t8))
# <class 'tuple'>
```

```python
t1 = (10,20,30,40,50)
print(t1[0])
# 10
print(t1[-5])
# 10
print(t1[0:3])
# (10, 20, 30)
print(t1[:2])
# (10, 20)
print(t1[2:])
# (30, 40, 50)
print(t1[0:4:2])
# (10, 30)
print(t1[::-2])
# (50, 30, 10)
```

```python
t1 = (10, 20, 30, 40, 50)
# t1[0] = 100
# TypeError: 'tuple' object does not support item assignment
```

```python
l1 = list(t1)
l1[0] = 100
t1 = tuple(l1)
print(t1)
# (100, 20, 30, 40, 50)
```

```python
t1 = (10, 20, 30, 40, 50)
e1, e2, e3, e4, e5 = t1
print(e1,e2,e3,e4,e5)
# 10 20 30 40 50
e1, *t2, e5 = t1
print(e1,t2,e5)
# 10 [20, 30, 40] 50
t2 = tuple(t2)
print(t2)
# (20, 30, 40)
e2, *t3 = t2
e3, e4 = t3
print(e2,e3,e4)
# 20 30 40
```
